```python
# import the necessary document

import pandas as pd
import numpy as np
```

```python
# creating a dataframe 

data = [
    [101, 'Ada',    'Paris',      'female', '$100,000', 'data scientist',   41, 88.0],
    [102, 'Ann',    'Toronto',    'female', '$95,000',  'digital marketer', 28, 79.0],
    [103, 'Chima',  'California', 'male',   '$120,000', 'data engineer',    33, 81.0],
    [104, 'Obi',    'Shanghai',   'male',   '$85,000',  'product manager',  34, 80.0],
    [105, 'Stacey', 'Manchester', 'female', '$97,000',  'ui/ux designer',   38, 68.0],
    [106, 'Ola',    'Cairo',      'female', '$98,000',  'software engineer',31, 61.0], 
    [107, 'Lyon',   'Osaka',      'male',   '$96,000',  'project manager',  37, 84.0],  
]

columns = ['ID', 'Name', 'City', 'Gender', 'Salary', 'Occupation', 'Age', 'Py-score']
df = pd.DataFrame(data, columns=columns)

```

```python
''' 
we are going to insert new column using diff method

'''

df['AWS-score'] = np.array([71.0, 95.0, 88.0, 79.0, 91.0, 91.0, 75.0])

df

```

```python
# adding a new column GCP-score with the .insert() method, using loc to identify the particular location in our datafarme

df.insert(loc= 5, column = 'GCP-score',
          value=np.array([80.0, 83.0, 78.0, 88.0, 75.0, 72.0, 81.0]))
df
```

```python
# adding a column at the end in your DataFrame directily without specify the location and a particular values

df['Bech-mark'] = 60.0
df
```

```python
'''
we can perfom basic arithemetics in our DF using our labels or column in our Dataframe

'''

df['Total score']= df['GCP-score'] + df['AWS-score']
print('Total score:')
print(df['Total score'])
```

```python
df['Py-score'] / df['Bech-mark']
print('Answer')
print(df['Py-score'])
```

```python
# we can use the basic arithemetics to add new column 

df['Total'] =\
0.4 * df['Py-score'] + 0.3 * df['GCP-score'] + 0.3 * df['AWS-score']

print('Newscore added')
print(df['Total'])
```
