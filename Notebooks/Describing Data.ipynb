# Determining Data Statistics
pandas provides many statistical methods for DataFrames. You can get basic statistics for the numerical columns of a pandas DataFrame with .describe():

```python
# To get the summary of our DataFrame we use .describe() but first let start by creating our dataset

import pandas as pd
import numpy as np

# creating a dataframe 

data = [
    [101, 'Ada',    'Paris',      'female', '$100,000', 'data scientist',   41, 88.0],
    [102, 'Ann',    'Toronto',    'female', '$95,000',  'digital marketer', 28, 79.0],
    [103, 'Chima',  'California', 'male',   '$120,000', 'data engineer',    33, 81.0],
    [104, 'Obi',    'Shanghai',   'male',   '$85,000',  'product manager',  34, 80.0],
    [105, 'Stacey', 'Manchester', 'female', '$97,000',  'ui/ux designer',   38, 68.0],
    [106, 'Ola',    'Cairo',      'female', '$98,000',  'software engineer',31, 61.0], 
    [107, 'Lyon',   'Osaka',      'male',   '$96,000',  'project manager',  37, 84.0],  
]

columns = ['ID', 'Name', 'City', 'Gender', 'Salary', 'Occupation', 'Age', 'Py-score']
df = pd.DataFrame(data, columns=columns)
```

```python
# to get the summary statistics like count, mean, std, min, etc of our dataframe

df.describe()
```

```python
# Recreating my DF and cleaning my salary column from str to float dtype

data = [
    [101, 'Ada',    'Paris',      'female', '$100,000', 'data scientist',   41, 88.0],
    [102, 'Ann',    'Toronto',    'female', '$95,000',  'digital marketer', 28, 79.0],
    [103, 'Chima',  'California', 'male',   '$120,000', 'data engineer',    33, 81.0],
    [104, 'Obi',    'Shanghai',   'male',   '$85,000',  'product manager',  34, 80.0],
    [105, 'Stacey', 'Manchester', 'female', '$97,000',  'ui/ux designer',   38, 68.0],
    [106, 'Ola',    'Cairo',      'female', '$98,000',  'software engineer',31, 61.0], 
    [107, 'Lyon',   'Osaka',      'male',   '$96,000',  'project manager',  37, 84.0],  
]
columns = ['ID', 'Name', 'City', 'Gender', 'Salary', 'Occupation', 'Age', 'Py-score']
df = pd.DataFrame(data, columns=columns)

# Clean the Salary column — remove $ and , then convert to integer
df['Salary'] = df['Salary'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(int)

# Now describe() will include all numeric columns
df.describe()
```

```python
df.mean(numeric_only = True)
```

```python
df.std(numeric_only =True)
```
