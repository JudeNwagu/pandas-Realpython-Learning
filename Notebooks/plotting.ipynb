## Plotting With pandas DataFrames

pandas allows you to visualize data or create plots based on DataFrames. It uses Matplotlib in the background, so exploiting pandas’ plotting capabilities is very similar to working with Matplotlib.

 To display the plots or any visualization, then you first need to import 
 
 - matplotlib.pyplot: the plotting library that pandas uses under the hood to draw
 - pandas: the main library for working with tables of data (DataFrames)
 - numpy: used for generating random numbers and math operations
 

```python
# import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# using our temperature dataset we are hoing to plot or visualize our dataset

# we create a variable temperature, which refers to the list of temperature values.
# This list contains 24 temperature readings, representing each hour of a day.

temp = [ 8.0,  7.1,  6.8,  6.4,  6.0,  5.4,  4.8,  5.0, 9.1, 12.8, 15.3, 19.1, 21.2, 22.1, 22.4, 23.1,
        21.0, 17.9, 15.5, 14.4, 11.9, 11.0, 10.2,  9.1]

temp = pd.Series(temp)

temp
```

## Now you can use pandas.DataFrame.plot() to create the plot and plt.show() to display it:

```python
temp.plot(kind='line', title="Temperature Trend")
plt.ylabel("Temperature")
plt.xlabel("Observation")
plt.show()
```

```python
# This code creates the plot and saves it as a file called 'temperatures.png' in your working directory.
plt.plot(temperature)
plt.savefig('temperatures.png')
```

```python
temp.plot(kind='bar', title="Temperature by Observation")
plt.ylabel("Temperature")
plt.xlabel("Observation")
plt.show()
```

```python
temp.plot(kind='hist', bins=8, title="Temperature Distribution")
plt.xlabel("Temperature")
plt.show()
```

```python
temp.plot(kind='box', title="Temperature Summary")
plt.show()
```

```python
df = pd.DataFrame({
    "Hour": range(24),
    "Temperature": temperature
})

df.plot(kind='scatter', x='Hour', y='Temperature', title="Temperature by Hour")
plt.show()
```
