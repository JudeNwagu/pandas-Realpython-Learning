# Creating a Pandas DataFrame Using A Dictionary

```python
# Creating a DF with Dict

data = {
    'Student': ['Ada', 'Chima', 'Obi', 'Lyon', 'Stacy'],
    'rank' :np.array([1, 4, 3, 5, 2]), 
    'Mark': 100
}


row_labels = [101, 102, 103, 104, 105,]

df = pd.DataFrame(data=data, index=row_labels)

# Display the dataframe

print(df)
```

```python
# dataset
data = [
    {'Student': 'Ada', 'rank' : 1, 'Mark': 100},
    {'Student': 'Chima', 'rank' : 2, 'Mark': 100},
    {'Student': 'Obi', 'rank' : 3, 'Mark': 100},
    {'Student': 'Lyon', 'rank' : 4, 'Mark': 100},
    {'Student': 'Stacy', 'rank' : 5, 'Mark': 100}
]

# Create the nested list 'l' using list comprehension
l = [[item['Student'], item['rank'], item['Mark']] for item in data]

# Display the nested list
print("Nested List (l):", l)

# Create the DataFrame using the nested list
df = pd.DataFrame(l, columns=['Student', 'rank', 'Mark'])

print("\nDataFrame from nested list:\n", df)
```

# Creating a pandas DataFrame With NumPy Arrays

```python
# Extract each values from our dictionary 

arr = np.array([
    [d["Student"], d["rank"], d["Mark"]]
    for d in data
])
```

```python
#View array we just extracted
arr
```

```python
# converting it back to a dataframe

df = pd.DataFrame(arr, columns=["Student", "Rank", "Mark"])

df
```

```python
# splitting the data type to get only the numeric data

arr = np.array([
    [d["rank"], d["Mark"]]
    for d in data
])

arr
```

```python
df = pd.DataFrame(arr,columns=["rank","Mark"])

df["Student"] = [d["Student"] for d in data]

df
```

# Creating a pandas DataFrame From Files

```python
'''
You can save and load the data and labels from a pandas DataFrame to and from a number of file types, 
including CSV, Excel, SQL, JSON, and more. This is a very powerful feature.
'''

# Save this raw data as a csv file using to_csv()

# Create the DataFrame from the raw data
data = [
    [101, 'Ada', 'Paris', 'female', '$100,000','data scientist', 41, 88.0,],
    [102, 'Ann', 'Toronto', 'female', '$95,000','digital markter',  28, 79.0],
    [103, 'Chima', 'California', 'male','$120,000','data engineer', 33, 81.0],
    [104, 'Obi', 'Shanghai', 'male', '$85,000','product manager', 34, 80.0],
    [105, 'Stacey', 'Manchester', 'female','$97,000','ui/ux designer', 38, 68.0],
    [106, 'Ola', 'Cairo', '$98,000', 'female','software engineer', 31, 61.0],
    [107, 'Lyon', 'Osaka', '$96,000', 'male','project manager', 37, 84.0]
]

# Define column names
columns = ['', 'Name', 'City', 'Gender', 'Salary','Occupation', 'Age', 'Py-score']

# Create DataFrame
df = pd.DataFrame(data, columns=columns)

# Save as CSV file
df.to_csv('data.csv', index=False)

```

# Accessing and Modifying Data

```python
'''
In pandas you can access data using the four accessor in the following way:

1. .loc[] accepts the labels of rows and columns and returns Series or DataFrames. 
You can use it to get entire rows or columns, as well as their parts.

2. .iloc[] accepts the zero-based indices of rows and columns and returns Series or DataFrames. 
You can use it to get entire rows or columns, or their parts.

3. .at[] accepts the labels of rows and columns and returns a single data value.

4. .iat[] accepts the zero-based indices of rows and columns and returns a single data value.

Note: .loc[] and .iloc[] are particularly powerful. They support slicing and NumPy-style indexing. 
You can use them to access a column in the following way:

'''
```

```python
# df.loc[:, 'Occupation'] returns the column city. The slice construct (:) in the row label place means that all the rows should be included.
df.loc[:, 'Occupation']
```

```python
# df.iloc[:, 4] returns the same column because the zero-based index 1 refers to the four column, occupation.

df.iloc[:, 4]
```

```python
df.iloc[1:6:2, 0]
```

```python
df.iloc[1:6, [0,1]]
```
